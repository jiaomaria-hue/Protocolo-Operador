ainda na aula 7: tambem ele explica sobre:  professor explica que, dentro das chaves {} do .format(), você pode definir um "espaço reservado" (por exemplo, 20 caracteres) e um caractere de controle para definir o alinhamento:

Alinhamento à Direita (>): O conteúdo é movido para o final do espaço definido. Exemplo: {:<20}.
Alinhamento à Esquerda (<): O conteúdo fica no início do espaço definido. Exemplo: {:>20}.
Centralização (^): O conteúdo é colocado exatamente no meio do espaço definido. Exemplo: {:^20}.
Uso de caracteres de preenchimento:
Além de definir o lado, você pode colocar um caractere antes do símbolo de alinhamento para preencher o "vazio" restante. Por exemplo, ao usar {:=>20}, o Python preencherá o espaço livre com o sinal de igual (=) em vez de deixar espaços em branco.

 EXEMPLO: print('Olá, {:>20}!'.format(nome))
